```python
# replace T with U in lncRNA transcripts
from Bio import SeqIO
import os

data_path = '/home/anna/medved/ceRNAs/ceRNAs_tests/data/'
lncrnas_path = os.path.join(data_path, 'lncRNAs.fasta')
out_path = os.path.join(data_path, 'lncRNAs_U.fasta')

output = open(out_path,'a') 
for record in SeqIO.parse(lncrnas_path, "fasta"):
    rna_seq = str(record.seq).replace('T', 'U')
    output.write(">%s\n%s\n"%(record.description, rna_seq))
output.close()
```

```python
precursor_path = os.path.join(data_path, 'precursor.hs.mirnas.fa')
lncrnas_path = os.path.join(data_path, 'lncRNAs.fasta')
out_path = os.path.join(data_path, 'lncrna_mir_precursors.txt')

output = open(out_path,'a') 
for lncrna in SeqIO.parse(lncrnas_path, "fasta"):
    for mir_precursor in SeqIO.parse(precursor_path, "fasta"):
        if str(mir_precursor.seq) in str(lncrna.seq):
            print(lncrna.id, mir_precursor.id)
            output.write("%s\t%s\n"%(lncrna.id, mir_precursor.id))
output.close()
```
