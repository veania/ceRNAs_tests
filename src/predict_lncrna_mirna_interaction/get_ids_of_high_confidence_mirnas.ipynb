```python

from Bio import SeqIO
import os
from __future__ import print_function

fasta_path = os.path.join(os.getcwd(), '../../data/mature_high_conf_v22_1.fa')
high_conf_mirnas_list_path = os.path.join(os.getcwd(), '../../data/mature_high_conf_v22_1.txt')
```

```python
output = open(high_conf_mirnas_list_path,'a') 
for record in SeqIO.parse(fasta_path, "fasta"):
    if 'hsa' in record.description:
        id_formatted = record.description.split(' ')[0]
        print(id_formatted)
        output.write("%s\n"%(id_formatted))
output.close()
#     if id_formatted in target_ids:
#         rna_seq = str(record.seq).upper()
```
